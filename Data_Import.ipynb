## Calcuations required
### Ocupation mix
Occupation mix controls for differences in the composition of jobs across boroughs, ensuring that observed wage differences are not simply driven by types of jobs present. The aim is to seperate wage inflation (London pays more for the same job) factors from job mix (London has more high paying jobs).

Question: Are people earning more because of where they are, or because of what jobs exist there?

Used to:
- Adjust London wage comparisons
- Compare: inner London (high-skill concentration) to outer London (more service jobs)

#### Approach: loop for each Borough
1. Agregate the number of jobs within each of the SOC codes per borough by the following Skill classifications:
| SOC Code | Group / Skill | Skill Classification |
|----------|--------|--------|
| 1 | Managers, directors, senior officials | High |
| 2 | Professional occupations | High | 
| 3 | Associate professional & technical | High |
| 4 | Administrative & secretarial | Mid |
| 5 | Skilled trades | Mid |
| 6 | Caring, leisure & service | Low | 
| 7 | Sales & customer service| Low | 
| 8 | Process, plant & machine operatives | Low |
| 9 | Elementary occupations | Low| 

2. Create Skill Mix indicators
- share_high = high_skill_jobs / total_jobs
- share_mid  = mid_skill_jobs / total_jobs
- share_low  = low_skill_jobs / total_jobs

### Household mix
To weight the scoial participation costs to cover all family memebers / dependents.
hypotheisis: London workers have fewer dependents, and therefore could contribute to higher disposable income.


## API 
### Nomis 
Nomis has a proper REST API that supports discovery + data download. Using API to pull borough-level (32):
- earnings
- employment by sector
- occupations

Can also return CSV, JSON, Excel directly.


## Manual Download
### ASHE - static input
Public data is available as aggregated tables (CSV / Excel) for download. Subset by London:
- Wage levels 
- Occupation splits

### Land Registry - static input
Download borough-level house price index OR aggregated London dataset:
- Price Paid Data (CSV)
- UK House Price Index (CSV)
